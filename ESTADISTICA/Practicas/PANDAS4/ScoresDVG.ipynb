### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

### Step 2. Create the DataFrame it should look like below.

```python
df = pd.DataFrame()
df['first_name'] = ['Jason', 'Molly', 'Tina', 'Jake', 'Amy']
df['last_name'] = ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze']
df['age'] = [42, 52, 36, 24, 73]
df['female'] = [0, 1, 1, 0, 1]
df['preTestScore'] = [4, 24, 31, 2, 3]
df['postTestScore'] = [25, 94, 57, 62, 70]
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
sns.scatterplot(x="preTestScore", y="postTestScore", size="age",
            sizes=(30, 200), alpha=0.85, palette="muted", data=df, legend=False)
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
df.rename(columns={"female": "sex"}, inplace=True)
df['sex'].loc[df['sex'] == 0] = "Male"
df['sex'].loc[df['sex'] == 1] = "Female"
```

```python
sns.relplot (x="preTestScore", y="postTestScore", s=df['postTestScore']*4.5, hue="sex", data=df)
```


- [Cows_and_Bulls]

Say the number generated by the computer is 1038. An example interaction could look like this:

  Welcome to the Cows and Bulls Game! 
  
  >Enter a number:

  >> 1234

  >>> 2 cows, 0 bulls

  >> 1253

  >>> 1 cow, 1 bull
  
  ...
Until the user guesses the number or write "STOP"

```python
def CandB():
    win = input("introduce la contraseña a adivinar, 4 cifras")
    num = input("introduce un número")
    cows = 0
    bulls = 0
    if len(win) != 4:
        print("Debes escribir una contraseña de 4 cifras")
    if len(num) != 4:
        print("Debes introducir un número de 4 cifras")
    elif len(num) == 4:
        for pos, val in enumerate(win):            
            for pos2, val2 in enumerate(num):
                if val2 in win and pos != pos2:
                        bulls += 1
                if pos == pos2 and val == val2:
                        cows += 1
                                  
                if cows == 4:
                    print("Has ganado! La contraseña era: ", win)
            
                
            return "Cows: ", cows, "Bulls :", bulls
    

            
    
```

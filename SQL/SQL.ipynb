```
import sqlite3
import pandas as pd
```

```
dir_db = "chinook.db"
```

```
conn = sqlite3.connect(dir_db)
```

```
cursor = conn.cursor()
```

```
query = """
SELECT *
FROM customers, invoices
WHERE customers.CustomerId = i.CustomerId
"""

query = """
SELECT c.FirstName, c.LastName, i.InvoiceId, i.InvoiceDate, i.BillingCountry
FROM customers c
JOIN invoices i ON c.CustomerId = i.CustomerId
WHERE c.Country = 'Brazil'
"""     #son 2 querys equivalentes. Primero se ejecutan FROM y JOIN, luego SELECT. Por eso peudes dar el alias "c" a customers y sirve 
```

```
pd.read_sql_query(query, conn)
```

```
query2 = """
SELECT i.*
FROM customers c
JOIN invoices i ON c.CustomerId = i.CustomerId
WHERE c.Country = 'Brazil'
"""
```

```
pd.read_sql_query(query2, conn)
```

```
query3 = """
SELECT i.*, e.FirstName || ' ' || e.LastName as 'nombre empleado'
FROM employees e
JOIN customers c ON e.EmployeeId = c.SupportRepId
JOIN invoices i on c.CustomerId = i.CustomerId
"""

df = pd.read_sql_query(query3, conn)
df
```

```
query8 = """
SELECT e.FirstName, e.LastName, COUNT(*) InvoiceId
FROM employees e
JOIN customers c ON e.EmployeeId = c.SupportRepId
JOIN invoices i on c.CustomerId = i.CustomerId
GROUP BY e.EmployeeId
"""

df = pd.read_sql_query(query8, conn)
df
```
